```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

def load():
    path="serie_historica_MD.csv"
    data=pd.read_csv(path,delimiter=',')
    #print(data)
    return data

def main():
    df=load()
    print(df)
    hospitalized_count=df.groupby('CCAA')['Hospitalizados'].nunique()
    print(hospitalized_count)
    
    print(df.info())
    null_columns=df.isnull().any()
    print(null_columns)
    print(df.isnull().sum())
    
    df['FECHA']=pd.to_datetime(df['FECHA'])
    df=df.sort_values(by='FECHA')
    
    
    df_ct=df[df['CCAA']=='CT'].copy()
    print(df_ct)
    
    df_ct=df_ct.sort_values(by='FECHA')
    
    missing=df_ct['Hospitalizados'].isnull().sum()
    print("faltantes ",missing)
    
    """plt.figure(figsize=(10,6))
    plt.plot(df_ct['FECHA'],df_ct['Hospitalizados'],marker='x',label='Datos originales')
    plt.show()"""
    
    df_ct['Hospitalizados']=df_ct["Hospitalizados"].fillna(df_ct['Hospitalizados'].mean())
    missing_d=df_ct["Hospitalizados"].isnull().sum()
    print("faltantes ",missing_d)

main()
```
